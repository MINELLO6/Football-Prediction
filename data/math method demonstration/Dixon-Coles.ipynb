## Models

### Maher's Model (1982)

Let $X_{i,j}$ and $Y_{i,j}$ denote the number of goals scored by the home and away teams, respectively, in a match between teams $i$ and $j$. 

#### Assumptions:
- **Independence**: The number of goals scored by the home and away teams in any particular game are independent Poisson variables, determined by the respective attack and defense qualities of each side. That is, $X_{i,j}$ and $Y_{i,j}$ are independent.
- **Home Advantage**: A parameter $\gamma > 0$ accounts for the home team's advantage.
- **Attack and Defense Factors**: Each team $i$ has an attack strength $\alpha_i$ and a defense strength $\beta_i$, both of which are positive.

#### Mathematical Formulation:
The goal counts follow Poisson distributions:

$$
X_{i,j} \sim \text{Poisson}(\alpha_i \beta_j \gamma)
$$

$$
Y_{i,j} \sim \text{Poisson}(\alpha_j \beta_i)
$$

where:
- $\alpha_i > 0$ represents the attack strength of team $i$.
- $\beta_i > 0$ represents the defense strength of team $i$.
- $\gamma > 0$ accounts for the home advantage.

Maher (1982) also introduced a more general model that allows for separate home and away attack and defense parameters for each team. However, this formulation serves as a useful simplification.


## Models

### Dixon and Coles Model (1997)
https://www.ajbuckeconbikesail.net/wkpapers/Airports/MVPoisson/soccer_betting.pdf

Maher (1982) proposed an independent Poisson model for football scores, but this assumption of independence has some limitations. Dixon and Coles (1997) introduced a modification to better capture **the dependence between low-scoring events**, which Maher's model fails to account for.

#### **Mathematical Formulation**
Instead of assuming independence between home and away goal counts, Dixon and Coles proposed the following modification:

$$
\Pr(X_{i,j} = x, Y_{i,j} = y) = \tau_{\lambda, \mu}(x, y) \frac{\lambda^x e^{-\lambda}}{x!} \cdot \frac{\mu^y e^{-\mu}}{y!}
$$

where:
- $\lambda = \alpha_i \beta_j \gamma$ represents the expected number of goals scored by the home team.
- $\mu = \alpha_j \beta_i$ represents the expected number of goals scored by the away team.

#### **Dependence Adjustment**
To account for dependence in low-scoring games, a correction factor $\tau_{\lambda, \mu}(x, y)$ is introduced:

$$
\tau_{\lambda, \mu}(x, y) =
\begin{cases}
1 - \lambda \mu \rho, & \text{if } x = 0, y = 0, \\
1 + \lambda \rho, & \text{if } x = 0, y = 1, \\
1 + \mu \rho, & \text{if } x = 1, y = 0, \\
1 - \rho, & \text{if } x = 1, y = 1, \\
1, & \text{otherwise}.
\end{cases}
$$

where $\rho$ is the **dependence parameter** that adjusts the probability for low-scoring outcomes. 

#### **Interpretation of $\rho$**
- If $\rho = 0$, the model reduces to the independent Poisson model.
- If $\rho \neq 0$, the probabilities of low-scoring games (e.g., $0$-$0$, $1$-$0$, $0$-$1$, $1$-$1$) are adjusted.
- The range of $\rho$ is constrained by:

$$
\max(-1/\lambda, -1/\mu) \leq \rho \leq \min(1/\lambda \mu, 1).
$$

#### **Key Features**
- **Improves fit for low-scoring matches**: The correction factor $\tau_{\lambda, \mu}(x, y)$ modifies the independent Poisson probabilities to better reflect real-world football results.
- **Maintains Poisson marginals**: The marginal distributions for home and away goals remain Poisson with means $\lambda$ and $\mu$ respectively.
- **Static Model**: Attack and defense parameters are assumed to be constant over time, which may limit predictive performance in dynamic settings.

Dixon and Coles (1997) demonstrated that this **dependence-adjusted model** provides a better fit for football match results compared to Maher's independent Poisson model, particularly for low-scoring games.
